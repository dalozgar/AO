```python
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 21 11:15:10 2021

@author: dlgma
"""

import numpy as np
 
class Graph():
 
    def __init__(self, vertices):
        self.V = vertices
        self.ady_matrix = np.array([[np.inf for column in range(vertices)]
                    for row in range(vertices)])
        for i in np.arange(self.V):
          self.ady_matrix[i][i]=0
        self.src = -1
        self.dist = np.array([np.inf] * self.V)
        self.parents = np.array([-1] * self.V )
        self.sptSet = np.array([False] * self.V)


    def set_edge (self, src, dest, weight):
      self.ady_matrix[src][dest] = weight

    #Esta función encuentra el siguiente arco a incluir en el
    # grafo de expansión (spanning tree). Devuelve el siguiente vertice
    # a procesar
    def get_next_vertex (self,ini):
      
      sig = -1
      
      #actualizo distancias       
      for index, distance in enumerate (self.ady_matrix[ini]):
        
        if (self.sptSet[index]==False):
        
          if (self.dist[index] > distance+self.dist[ini]): 
              self.dist[index] = distance +self.dist[ini]
              self.parents[index] = ini
              
     
      #descubro el siguiente
      dist_sig = np.inf
      for i, value in enumerate (self.sptSet):
         
          if (value == False):
               dist_i = self.dist[i]
               if (dist_sig > dist_i):
                   dist_sig = dist_i
                   sig = i
              
      if (sig != -1):
          self.sptSet[sig] = True
         
      return sig


    def dijkstra(self, src):
 
        self.src = src
        self.dist[src] = 0
        self.parents[src] = src
        self.sptSet[src] = True

        ini = src
      
        for i in range(self.V):
          ini = self.get_next_vertex (ini)
          
    # Esta funcion devuelve la ruta de origen a destino en forma 
    # de lista de vertices, en orden inverso
    def get_route(self,dest):
        
        ruta = np.array([])
        index = dest
        destino_inalcanzable = False
        while (index != self.src):
            if (self.dist[index]!=np.inf):
                ruta = np.append(ruta, index)
                
                index = self.parents[index]
            else:
                destino_inalcanzable = True
                print("WARNING: destino inalcanzable")
                break
             
        if not destino_inalcanzable:
            ruta = np.append(ruta, index)
        return ruta

#######################

g = Graph(7)

g.set_edge(0,1,5)
g.set_edge(0,2,4)
g.set_edge(0,3,3)
g.set_edge(1,3,2)
g.set_edge(1,4,3)
g.set_edge(1,6,11)
g.set_edge(2,3,1)
g.set_edge(2,5,4)
g.set_edge(2,6,10)
g.set_edge(4,6,4)
g.set_edge(5,6,3)
g.set_edge(3,4,5)
g.set_edge(3,5,6)
g.set_edge(3,6,9)


g.dijkstra(0)

dest = 6
ruta = g.get_route(dest)
print(f"ruta a {dest} = {ruta}")
```
